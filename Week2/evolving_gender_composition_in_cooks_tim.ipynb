## 1. What is an interesting insight you found? 
`When, where, who, what, why, how.`

From 1860 to 1940, the gender composition of cooks in New York City had gradually transition from female-majority to male-majority. The reversed gender composition can potentially be attributed to the shifting ethnic composition of the cooks, among other reasons such as drop in the number of domestic workers. 



#### Visualization of the gender composition of cooks (1860 - 1940)

```python
year_list = [1860,1870,1880,1900,1910,1920,1930,1940]
```

```python
for year in year_list:
  describe('Gender', eval('cook_'+str(year)), plot_type = 'pie', year = year)
```
